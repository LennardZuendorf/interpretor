# Interpretor Data Set

Notebook used to edit the existing dataset for usage in the interpretor projects.

```python
from datasets import load_dataset

raw_dataset=load_dataset("LennardZuendorf/Dynamically-Generated-Hate-Speech-Dataset", "Dynamically Generated Hate Dataset v0.2.3.csv", split="train")
```

Editing Columns of Dynamically-Generated-Hate-Speech-Dataset, generating training and test split 

```python
dataset = raw_dataset.rename_column("Unnamed: 0", "id")
dataset = dataset.remove_columns(["acl.id", "X1", "type", "target", "level","split", "round.base","annotator","round","acl.id.matched"])
dataset = dataset.train_test_split(test_size=0.1)

print(dataset)
```

uploading new dataset into different repository

```python
dataset.push_to_hub("LennardZuendorf/interpretor-data")
```
